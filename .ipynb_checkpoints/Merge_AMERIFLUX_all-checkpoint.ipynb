```python
def convert_timezone_gmt(df,state):
    if state=="WA" or state=="OR" or state=="NV"or state=="CA":
        tz = pytz.timezone('US/Pacific')
    #Still more states to add
    elif state=="AZ" or state=="WY" or state=="UT" or state=="CO" or state=="NM" or state=="ID": 
        tz = pytz.timezone('US/Mountain')
    elif state=="OK" or state=="AR" or state=="IL" or state=="ND" or state=="WI" or state=="AL" \
    or state=="KS" or state=="MO" or state=="TX" or state=="NE" or state=="MN":
        tz=pytz.timezone('US/Central')
    else:
        tz=pytz.timezone('US/Eastern')
    df["Datetime_Local"]=df["Datetime"].dt.tz_localize(tz, nonexistent='shift_forward',ambiguous=False)
    df["Datetime_GMT"]=df["Datetime_Local"].dt.tz_convert('GMT')
    return df
## Change local timezones to GMT for merging 
all_st_p2=[]
for i in range(len(am_flux)):
    all_st_p2.append(convert_timezone_gmt(am_flux[i],am_flux[i]["State"].iloc[0]))
```

```python
## Creat a time variable fr
def get_coinciding_amflux_data(df_am,df_ls):
    df_ls["Datetime_Local"]=df_ls["date"].dt.tz_localize("GMT", nonexistent='shift_forward',ambiguous=False)
    df_ls["Datetime_GMT"]=df_ls["Datetime_Local"].dt.round("30min")
    df_merge=pd.merge(df_ls,df_am,on="Datetime_GMT",how="left")
    return df_merge
```

```python
def MAPE(Y_actual,Y_Predicted):
    mape = np.mean(( Y_Predicted-Y_actual)/Y_actual)*100
    return mape
from sklearn.metrics import mean_squared_error 
```

```python
def rmse(Y_actual,Y_Predicted):
    return mean_squared_error(Y_actual,Y_Predicted,squared=False)
```

```python
## Call errors 
netrad_error=[]
h_error=[]
le_error=[]
g_error=[]
for i in range(len(merged)):
    if merged[i][merged[i]["LEinst"]>0].shape[0]==0:
        netrad_error.append(np.nan)
        h_error.append(np.nan)
        le_error.append(np.nan)
        g_error.append(np.nan)
    else:
        netrad_error.append(MAPE(merged[i][(merged[i]["Rn"].notna()) & (merged[i][netrad_col[i]].notna())][netrad_col[i]],\
                                merged[i][(merged[i]["Rn"].notna()) & (merged[i][netrad_col[i]].notna())]["Rn"]))
        h_error.append(MAPE(merged[i][(merged[i]["Hinst"]>-10) & (merged[i]["Hinst"]<1000) & (merged[i]["Hinst"].notna()) & (merged[i][h_col[i]].notna())][h_col[i]],\
                       merged[i][(merged[i]["Hinst"]>-10) & (merged[i]["Hinst"]<1000) & (merged[i]["Hinst"].notna()) & (merged[i][h_col[i]].notna())]["Hinst"]))
        le_error.append(MAPE(merged[i][(merged[i]["LEinst"]>0) & (merged[i]["LEinst"].notna()) & (merged[i][le_col[i]].notna())][le_col[i]],\
                       merged[i][(merged[i]["LEinst"]>0) & (merged[i]["LEinst"].notna()) & (merged[i][le_col[i]].notna())]["LEinst"]))  
        g_error.append(MAPE(merged[i][(merged[i]["Ginst"]>0) & (merged[i]["Ginst"].notna()) & (merged[i][g_col[i]].notna())][g_col[i]],\
                       merged[i][(merged[i]["Ginst"]>0) & (merged[i]["Ginst"].notna()) & (merged[i][g_col[i]].notna())]["Ginst"]))  
```

```python
netrad_rmse=[]
h_rmse=[]
le_rmse=[]
g_rmse=[]

for i in range(len(merged)):
    if merged[i][merged[i]["LEinst"]>0].shape[0]==0:
        netrad_rmse.append(np.nan)
        h_rmse.append(np.nan)
        le_rmse.append(np.nan)
        g_rmse.append(np.nan)
    else:
#         print(i)
#         print("LE",merged[i][le_col[i]].shape[0])
#         print("Missing",merged[i][le_col[i]].isna().sum())
#         print("H",merged[i][h_col[i]].shape[0])
#         print("Missing",merged[i][h_col[i]].isna().sum())
#         print("Netrad",merged[i][netrad_col[i]].shape[0])
#         print("Missing",merged[i][netrad_col[i]].isna().sum())
        netrad_rmse.append(rmse(merged[i][(merged[i]["Rn"].notna()) & (merged[i][netrad_col[i]].notna())][netrad_col[i]],\
                                merged[i][(merged[i]["Rn"].notna()) & (merged[i][netrad_col[i]].notna())]["Rn"]))
        h_rmse.append(rmse(merged[i][(merged[i]["Hinst"]>-10) & (merged[i]["Hinst"]<1000) & (merged[i]["Hinst"].notna()) & (merged[i][h_col[i]].notna())][h_col[i]],\
                       merged[i][(merged[i]["Hinst"]>-10) & (merged[i]["Hinst"]<1000) & (merged[i]["Hinst"].notna()) & (merged[i][h_col[i]].notna())]["Hinst"]))
        le_rmse.append(rmse(merged[i][(merged[i]["LEinst"]>0) & (merged[i]["LEinst"].notna()) & (merged[i][le_col[i]].notna())][le_col[i]],\
                       merged[i][(merged[i]["LEinst"]>0) & (merged[i]["LEinst"].notna()) & (merged[i][le_col[i]].notna())]["LEinst"]))  
        g_rmse.append(rmse(merged[i][(merged[i]["Ginst"]>0) & (merged[i]["Ginst"].notna()) & (merged[i][g_col[i]].notna())][g_col[i]],\
                       merged[i][(merged[i]["Ginst"]>0) & (merged[i]["Ginst"].notna()) & (merged[i][g_col[i]].notna())]["Ginst"]))  
        print(merged[i][g_col[i]].describe())        
#     print(merged[i][merged[i]["LEinst"]>0].shape[0])
#     print(merged[i][(merged[i]["Hinst"]>-10) & (merged[i]["Hinst"]<1000)].shape[0])
 
```

```python
lat=[]
lon=[]
mp=[]
mat=[]
clm=[]
elev=[]
veg=[]
for i in range(len(merged)):
    if merged[i][merged[i]["LEinst"]>0].shape[0]==0:
        lat.append(np.nan)
        lon.append(np.nan)
        mp.append(np.nan)
        mat.append(np.nan)
        clm.append(np.nan)
        elev.append(np.nan)
        veg.append(np.nan)
    else:
        lat.append(merged[i][merged[i]["latitude"].notna()]["latitude"].iloc[0])
        lon.append(merged[i][merged[i]["longitude"].notna()]["longitude"].iloc[0])
        mp.append(merged[i][merged[i]["MAP (mm)"].notna()]["MAP (mm)"].iloc[0])
        mat.append(merged[i][merged[i]["MAT (°C)"].notna()]["MAT (°C)"].iloc[0])
        clm.append(merged[i][merged[i]["Clim"].notna()]["Clim"].iloc[0])
        elev.append(merged[i][merged[i]["Elevation"].notna()]["Elevation"].iloc[0])
        veg.append(merged[i][merged[i]["Veg"].notna()]["Veg"].iloc[0])
    
```

```python
netrad_mean=[]
h_mean=[]
le_mean=[]
g_mean=[]

for i in range(len(merged)):
    if merged[i][merged[i]["LEinst"]>0].shape[0]==0:
        netrad_mean.append(np.nan)
        h_mean.append(np.nan)
        le_mean.append(np.nan)
        g_mean.append(np.nan)
    else:
#         print(i)
#         print("LE",merged[i][le_col[i]].shape[0])
#         print("Missing",merged[i][le_col[i]].isna().sum())
#         print("H",merged[i][h_col[i]].shape[0])
#         print("Missing",merged[i][h_col[i]].isna().sum())
#         print("Netrad",merged[i][netrad_col[i]].shape[0])
#         print("Missing",merged[i][netrad_col[i]].isna().sum())
        netrad_mean.append(merged[i][(merged[i]["Rn"].notna()) & (merged[i][netrad_col[i]].notna())][netrad_col[i]].mean())
        h_mean.append(merged[i][(merged[i]["Hinst"]>-10) & (merged[i]["Hinst"]<1000) & (merged[i]["Hinst"].notna()) & (merged[i][h_col[i]].notna())][h_col[i]].mean())
        le_mean.append(merged[i][(merged[i]["LEinst"]>0) & (merged[i]["LEinst"].notna()) & (merged[i][le_col[i]].notna())][le_col[i]].mean())
        g_mean.append(merged[i][(merged[i]["Ginst"]>0) & (merged[i]["Ginst"].notna()) & (merged[i][g_col[i]].notna())][g_col[i]].mean())
        print(merged[i][g_col[i]].describe())        
#     print(merged[i][merged[i]["LEinst"]>0].shape[0])
#     print(merged[i][(merged[i]["Hinst"]>-10) & (merged[i]["Hinst"]<1000)].shape[0])
 
```
